```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Forbes Business of Basketball website
url = "https://www.forbes.com/nba-valuations/list/#header:position"
```

```python
# Specify the path to the ChromeDriver executable
chromedriver_path = "/opt/homebrew/bin/chromedriver"  

# Kept getting a deprecation error, so had to create a Service object
service = Service(chromedriver_path)
```

```python
# Initialize the WebDriver
driver = webdriver.Chrome(service=service)
```

```python
# Open the URL
driver.get(url)
```

```python
# Had to let content filled in by JavaScript load so this made the program sleep 
time.sleep(45) 
```

```python
# At first the DataFrame only showed the first 10 results. Had to put in code that would "scroll" down the page to scrape the rest of the content
for _ in range(3):  
    driver.find_element(By.TAG_NAME, "body").send_keys(Keys.PAGE_DOWN)
    time.sleep(5) 
```

```python
# Stored the page source in a variable after the JavaScript loaded the table
page_source = driver.page_source
```

```python
# Parsed the HTML content from the page source
soup = BeautifulSoup(page_source, "html.parser")
```

```python
# Reading the table into a DataFrame
forbeslist = pd.read_html(str(soup))
```

```python
# Printing the DataFrame 
if forbeslist:
    for i, df in enumerate(forbeslist):
        print("Table", i + 1)
        print(df)
else:
    print("No tables found")


# Close the WebDriver
driver.quit()
```
