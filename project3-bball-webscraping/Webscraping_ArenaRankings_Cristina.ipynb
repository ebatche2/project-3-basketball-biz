```python
columnsplit = arenarankings_df['location'].str.split(", ", expand=True)
arenarankings_df = arenarankings_df.assign(City=columnsplit[0],State=columnsplit[1])

arenarankings_df = arenarankings_df.drop(['location'], axis=1)


arenarankings_df.head(200)
```

```python
arenarankings_df['cost'] = arenarankings_df['cost'].str.replace('$', '')

arenarankings_df.head(200)
```

```python
arenarankings_df['cost  Per  Seat'] = arenarankings_df['cost  Per  Seat'].str.replace('$', '')

arenarankings_df.head(200)
```
